```python
import pandas as pd
import datetime as datetime
import glob as glob
```

```python
tcsvs = glob.glob('Temperature_Data/*.csv')
pcsvs = glob.glob('Precipitation_Data/*.csv')
```

```python
temp_tocat = []
for csv in tcsvs:
    temp_df = pd.read_csv(csv)
    if '/' in temp_df['Start Date'][0]:
        temp_df['Start Date'] = [datetime.date(int('20'+item.split('/')[2]),map(int,item.split('/'))[0],map(int,item.split('/'))[1]) for item in temp_df['Start Date']]
        temp_df['End Date'] = [datetime.date(int('20'+item.split('/')[2]),map(int,item.split('/'))[0],map(int,item.split('/'))[1]) for item in temp_df['End Date']]
    elif '-' in temp_df['Start Date'][0]:
        temp_df['Start Date'] = [datetime.date(map(int,item.split('-'))[0],map(int,item.split('-'))[1],map(int,item.split('-'))[2]) for item in temp_df['Start Date']]
        temp_df['End Date'] = [datetime.date(map(int,item.split('-'))[0],map(int,item.split('-'))[1],map(int,item.split('-'))[2]) for item in temp_df['End Date']]
    else:
        print 'sad reaccs'
    region = csv.split('_')[2].split('Africa')[0]
    temp_df.insert(0,'GenRegion',region)
    temp_tocat.append(temp_df)
```

```python
all_temp = pd.concat(temp_tocat)
all_temp.to_csv('Temperature_Data/All_Temp.csv')
all_temp
```
